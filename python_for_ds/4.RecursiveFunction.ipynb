### Recursion

A function calling itself is called as recursion

```python
# eg
# #python program to print factorial of a number using recurion

def factorial(n):
    """This returns the factorial of a number"""
    
#     one line code
    return 1 if(n == 1) else n * factorial(n-1)
#     if(n == 1):
#         return 1
#     return n * factorial(n-1)
    
print(factorial(5))
```

### Python program to return fibonacci sequence upto nth term using recursive function

```python
def fibonacciSeries(n):
    return 1 if(n <= 1) else fibonacciSeries(n-1) + fibonacciSeries(n-2)

print(fibonacciSeries(3))
```

```python
# print sequence of fibonacci series

def fibonacciSeries(n):
    print("Number:",n)
    return n if(n <= 1) else fibonacciSeries(n-1) + fibonacciSeries(n-2)

for i in range(10):
    print("Index ",i)
    print(fibonacciSeries(i))
```
