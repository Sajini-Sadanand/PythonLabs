```python
import numpy as np
import pylab as py
import scipy.stats as stats
import matplotlib as mp
```

```python
# N(0,1)

std_normal = np.random.normal(loc = 0, scale = 1, size = 1000)
# 0th to 100th percentile of std_normal

for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
# generate 100 samples from N(20,5)
measurements = np.random.normal(loc = 20, scale = 5, size = 100)

stats.probplot(measurements,dist="norm",plot=py)
py.show()
```

```python
# trying the same for 1000 values

measurements = np.random.normal(loc= 10,scale = 5, size = 1000)

stats.probplot(measurements,dist="norm",plot=py)
py.show()
```

```python
# Basically to me it looks like Q-Q plot is just a plot of --> subset of 2 distributions in sorted order (subset in the form of percentiles) .
# I just tried to plot sorted values of 2 normal distributions (having different set of parameters) and got a straight line.

std_normal1 = np.random.normal(loc = 0, scale = 1, size=1000)
std_normal2 = np.random.normal(loc = 50, scale = 10, size=1000)

x = np.sort(std_normal1)
y = np.sort(std_normal2)
mp.pyplot.plot(x,y)
mp.pyplot.show()
```

```python
# Another example with a non-normal distribution

data=np.random.poisson(2.5, 1000)
stats.probplot(data, dist='poisson', sparams=(1,), plot=py)
py.show()
```
